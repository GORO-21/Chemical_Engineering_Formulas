```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def onda_correlation(L, muL, rhoL, DL , alphat , dp,g = 9.81 ):
    """ L(float): liquid mass flux (kg/m^2 . s)
        muL (float): Liquid viscosity (Pa.s)
        rhoL (float): Liquid density (kg/m^3)
        DL (float): Diffusivity of solute in solid (m^2/s)
        alphat (float): Total specific surface area of packing (m^2/m^3)
        dp (float): Nominal Packing diameter (m)
        g: gravitational acceleration (9.81m/s^2)
        
        returns kL : Liquid side mass transfer coefficient (m/s)"""
   

    return 0.0051*((L/muL)**(2/3))*((muL/(rhoL*DL)) ** (-0.5))*((alphat*dp)**(0.4))*(rhoL/(muL*g))**((-1)/3)
```

```python
if __name__ == "__main__":

    L = 5
    muL = 0.001
    rhoL = 1000
    DL = 1.5e-9
    alphat = 200 
    dp = 0.025

    OC = onda_correlation(L, muL, rhoL, DL , alphat , dp, g = 9.81 )

    print(f" kL:{OC:.4f}")
    
```

```python
L = np.linspace(1,10,100)

k_L = 0.0051*((L/muL)**(2/3))*((muL/(rhoL*DL)) ** (-0.5))*((alphat*dp)**(0.4))*(rhoL/(muL*9.81))**((-1)/3)

plt.figure(figsize=(8,6))
plt.plot(L,k_L, color='purple', label='k_L vs L')
plt.title("Mass transfer Coefficients vs Liquid Mass Flux in packed column")
plt.xlabel('Liquid Mass Flux(kg/m^2.s)')
plt.ylabel('Mass Transfer Coefficient (m/s)')
plt.grid(True)
plt.legend()
plt.show
           
```
